```python
class Student:
    students = []
    count = 0
    def __init__(self,id,name,marks):
        self.roll_no = id
        self.name = name 
        self.marks = marks
        Student.students.append(self)
        
    def cal_result(self):
        self.per = sum (self.marks) / 5
        if self.per > 40 :
            self.result = "pass"
        else:
            self.result = "fail"
        
        
stud1 = Student(101,"ABC",[70,50,60,50,50])
stud2 = Student(102,"PQR",[30,20,30,30,55])
stud3 = Student(103,"XYZ",[70,80,80,90,50])


del stud1
print(stud1)
print(Student.students)
# students = [stud1,stud2,stud3]
for student in Student.students:
    student.cal_result()
    
for student in Student.students:
    if student.result == "pass":
        print(student.name,student.per)

for student in Student.students:
    if student.result == "fail":
        print(student.name,student.per)
        
        
print(Student.count)
Student.count = 100 
print(stud1.count)

print(stud1.__dict__)
stud1.count = 200
print(stud1.__dict__)
print(stud1.count)
print(Student.count)
```

```python
class User:
    def __init__(self,username,email):
        self.username = username
        self.email = email
        
    def get_details(self):
        return self.username,self.email
    
class Demo:
    def get_details(self):
        return "HI"
    
class UserProfile(Demo,User):
    def __init__(self,username,email,github,contact):
        super().__init__(username,email)
        self.github = github
        self.contact = contact
        
    def get_details(self):
        username,email = User.get_details(self)
        return username,email,self.github,self.contact
    
profile1= UserProfile("abc","abc@gmail.com","www.github.com/abc","7123456789")
print(profile1.get_details())
```

```python
def add_deco(func):    
    def new_func(num1,num2):
        if type(num1) == type(num2):
            return func(num1,num2)
        else:
            return func(str(num1),str(num2))
        
    return new_func

@add_deco
def add(num1,num2):
    return (num1 + num2)

print(add(10,10))
print(add("python","java"))
print(add(10,"Python"))
```

```python
s = "12-05-2019"
r = re.compile('^([0-9]{2})-([0-9]{2})-([0-9]{4})$')
s = re.search(r,s)
print(s.group(3))
```

```python
s = "12-05-2019"
r = re.compile('^(?P<day>\d{2})-(?P<month>[0-9]{2})-(?P<year>[0-9]{4})$')
s = re.search(r,s)
print(s.group(0))
```
